# Netflix_Exploratory Data Analysis

# Exploring the dataset

```python
df = pd.read_excel('NetflixOriginals.xlsx')
df
```

# Checking datatype 

```python
df.info()
```

# converting premiere datatype to datetime datatype

```python
df['Premiere'] = pd.to_datetime(df['Premiere'], dayfirst = True)
df
```

# Adding cloumns Date, Months, Year to the dataset for future analysis 

```python
df['Day'] = df['Premiere'].apply(lambda x : x.day)
months = {1:'Jan', 2:'Feb', 3:'March', 4:'Apr', 5:'May', 6:'Jun', 7:'July', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
df['Month'] = df['Premiere'].apply(lambda x : months[x.month])
df['Year'] = df['Premiere'].apply(lambda x : x.year)
```

```python
df.head()
```

# Checking for Null 

```python
df.isnull().sum()
```

# Checking for duplicates

```python
df.duplicated().unique()
```

# Distribution of numeric features

```python
plt.figure(figsize=(10,5))
plt.subplot(3,1,1)
sns.boxplot('IMDB Score', data = df)
plt.xlabel('IMDB Score', fontsize = 15)


plt.figure(figsize=(10,5))
plt.subplot(3,1,2)
sns.boxplot('Runtime', data = df)
plt.xlabel('Runtime', fontsize = 15)


plt.figure(figsize=(10,5))
plt.subplot(3,1,3)
sns.boxplot('Year', data = df)
plt.xlabel('Year', fontsize = 15)
plt.show()
```

From the above distribution, 

The first movie permiere in the netflix is from the year 2014,

The IMDB score minimum around 2 and maxmimum is around 9,

The maximum runtime is around 165 to 175.


# Checking for correlation between IMDB and Runtime

```python
plt.figure(figsize=(10,10))
sns.scatterplot(y = df['IMDB Score'], x = df['Runtime'])
plt.xlabel('Runtime', fontsize = 15)
plt.ylabel('IMDB Score', fontsize = 15)
plt.title('IMDB Score and Runtime', fontsize = 15)
plt.show()
```

From the above graph we found that there is no correlation between Imdb and Runtime

```python
df.corr()
```

The correlation score:

1 : Strongly and positively correlated (one increases, other also increases and vice versa)

0 : No correlation

-1 : Strongly and negetively correlated (one increases, other also decreases and vice versa)

The value for IMDB and Runtime is -0.040896 so clear that there is no relation between these two features

# seeing which genre has got high IMDB score

```python
len(df['Genre'].unique())
```

There are 115 genre are availabel so we are going see the top 5 genre based on imdb score

```python
df_genre = df.groupby(['Genre']).mean(['IMDB Score']).sort_values(by='IMDB Score', ascending = False).reset_index().iloc[:5,:]
df_genre
```

```python
plt.figure(figsize=(5,5))
plt.pie(x = df_genre['IMDB Score'], labels = df_genre['Genre'], radius = 2, autopct = '%0.2f%%', explode = [0.3,0,0,0,0])
plt.show()
```

People like Animation/Christmas/Comedy/Adventure genre movies

# The top 10 languages movies in Netflix

```python
df_lang = df.groupby('Language')['Title'].count().reset_index().sort_values(by='Title', ascending = False).iloc[:10,:]

plt.figure(figsize=(15,10))
sns.barplot(x = 'Language', y = 'Title', data = df_lang)
plt.xlabel('Language', fontsize = 15)
plt.ylabel('Number of Movies', fontsize = 15)
plt.title('Top 10 Language movies in Netflix', fontsize = 20)
plt.show()
```

From the above graph we can see that english movies are released mostly in netflix

# Let' s see in which month mostly the movie releases

```python
df_release = df.groupby('Month')['Title'].count().reset_index().sort_values(by='Title')
x = list(np.arange(0,100,10))

plt.figure(figsize = (10,10))
x = plt.gca()
sns.barplot(x = 'Title', y = 'Month', orientation = 'horizontal', data = df_release)
x.set_yticklabels(df_release['Month'],fontsize = 20)
plt.xlim(0,100)
plt.xlabel('Number of movie Release',fontsize = 15)
plt.ylabel('Months',fontsize = 15)
plt.title('Month and Premieres',fontsize = 15)
plt.show()
```

The more movies releases are in the month october and followed by april, i think reason can be the holidays during these months.

# let's see how many good movies are released in netflix

good movies should have IMDB Score 7 and above.i took based on my search, i also provided link for this

```python
Threshold = 7

df['Best_Movie'] = df['IMDB Score'].apply(lambda x : 1 if x > Threshold else 0)
```

0 : if the IMDB score less then 7

1 : if the IMDB score is higher then 7
